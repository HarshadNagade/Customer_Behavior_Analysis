```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')

df.columns
```

```python
# create new col age_group
labels=['Young Adult','Adult','Middle-aged','Senior']
df['age_group']=pd.qcut(df['age'],q=4, labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
# Correct dictionary and variable name
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

# Map correctly
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)

# Remove rows where mapping failed (NaN)
df = df.dropna(subset=['purchase_frequency_days'])

# Display result
df[['purchase_frequency_days', 'frequency_of_purchases']].head(10)
```

```python
from sqlalchemy import create_engine

# Step 1: Database connection details
username = "postgres"
password = "123"
host = "localhost"
port = "5432"  # default PostgreSQL port
database = "customer_behavior"  # the database you created in pgAdmin

# Create SQLAlchemy engine
engine = create_engine(f"postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}")

# Step 2: Load DataFrame into PostgreSQL
table_name = "customer"  # choose any table name
df.to_sql(table_name, engine, if_exists="replace", index=False)

print(f"✅ Data successfully loaded into table '{table_name}' in database '{database}'.")
```
